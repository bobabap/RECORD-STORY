### Feature Engineering

Relative Perfomance 라는 새로운 feature를 계산하세요.

이는 최근 1년치 매출액의 평균값을 기준으로 

- 10% 이상 -> `S`
- 5% 이상 -> `A`
- -5 ~ 5% -> `B`
- -5%이하 -> `C`
- 10%이하 -> `D`

라는 값을 갖는 feature입니다.

20년도 2분기에 해당하는 결과값은 `A`가 나와야합니다.

각각에 해당하는 등급이 나오기 위해서 필요한 매출액을 추가로 서술하세요.

````python
df_val = list(df['매출액'].values)
take_mean = int(df['매출액'].mean())
def RP(vals):
  val_score = []
  for i in vals:
      if take_mean+(take_mean*0.1) < int(i): # 10% 이상
          val_score += 'S'
      elif take_mean+(take_mean*0.05) <= int(i) < take_mean+(take_mean*0.1): # 5% 이상
          val_score += "A"
      elif take_mean+(take_mean*0.05) > int(i) >= take_mean-(take_mean*0.05): # -5% 이상, 5% 이하
          val_score += 'B'
      elif take_mean-(take_mean*0.05) > int(i) >= take_mean-(take_mean*0.1): # -5%이하
          val_score += 'C'
      elif take_mean-(take_mean*0.1) >= int(i): # 10% 이하
          val_score += 'D'
  return val_score
score = RP(df_val)
score
````

````python
df['Relative Perfomance'] = score
df
````
