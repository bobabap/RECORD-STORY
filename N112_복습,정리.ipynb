### Feature Engineering

Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

ì´ëŠ” ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ 

- 10% ì´ìƒ -> `S`
- 5% ì´ìƒ -> `A`
- -5 ~ 5% -> `B`
- -5%ì´í•˜ -> `C`
- 10%ì´í•˜ -> `D`

ë¼ëŠ” ê°’ì„ ê°–ëŠ” featureì…ë‹ˆë‹¤.

20ë…„ë„ 2ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” ê²°ê³¼ê°’ì€ `A`ê°€ ë‚˜ì™€ì•¼í•©ë‹ˆë‹¤.

ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ì„œ í•„ìš”í•œ ë§¤ì¶œì•¡ì„ ì¶”ê°€ë¡œ ì„œìˆ í•˜ì„¸ìš”.

````python
df_val = list(df['ë§¤ì¶œì•¡'].values)
take_mean = int(df['ë§¤ì¶œì•¡'].mean())
def RP(vals):
  val_score = []
  for i in vals:
      if take_mean+(take_mean*0.1) < int(i): # 10% ì´ìƒ
          val_score += 'S'
      elif take_mean+(take_mean*0.05) <= int(i) < take_mean+(take_mean*0.1): # 5% ì´ìƒ
          val_score += "A"
      elif take_mean+(take_mean*0.05) > int(i) >= take_mean-(take_mean*0.05): # -5% ì´ìƒ, 5% ì´í•˜
          val_score += 'B'
      elif take_mean-(take_mean*0.05) > int(i) >= take_mean-(take_mean*0.1): # -5%ì´í•˜
          val_score += 'C'
      elif take_mean-(take_mean*0.1) >= int(i): # 10% ì´í•˜
          val_score += 'D'
  return val_score
score = RP(df_val)
score
````

````python
df['Relative Perfomance'] = score
df
````
